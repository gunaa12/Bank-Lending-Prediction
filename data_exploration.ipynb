```python
import pandas as pd
```

```python
train_dataset = pd.read_csv("data/train_preprocessed_data.csv")
train_dataset_loan_paid = train_dataset['loan_paid']
train_dataset = train_dataset.drop(columns=['loan_paid'])
train_dataset
```

```python
val_dataset = pd.read_csv("data/predict_preprocessed_data.csv")
print(val_dataset.columns)
val_dataset = val_dataset.drop(columns=['ID'])
val_dataset
```

```python
from scipy.stats import ttest_ind

significance_level = .4
mappings = []
for col in train_dataset.columns:
    stat, p_value = ttest_ind(train_dataset[col], val_dataset[col])
    if (p_value < significance_level):
        mappings.append((p_value, col, stat))

mappings = sorted(mappings)
for item in mappings:
    print(f"Col: {item[1]}, t-test: statistic={item[2]:.4f}, p-value={item[0]:.4f}")
```
